```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)      # –æ–∫—Ä—É–≥–ª—è–µ–º –¥–æ 1 –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
corr_matrix[np.abs(corr_matrix) < 0.3] = 0  # —É–±–∏—Ä–∞–µ–º —Å–ª–∏—à–∫–æ–º –º–∞–ª–µ–Ω—å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
corr_matrix
```

```python
plt.figure(figsize=(15, 12))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```
